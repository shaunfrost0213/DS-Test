### IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
```

### CREATING PANDAS DATAFRAME FOR TRAIN AND TEST DATASET

```python
X_train = pd.read_csv('Training/X_train.csv')
X_test =pd.read_csv('Test/X_test.csv')
Y_train = pd.read_csv('Training/y_train.csv')
```

```python
# printing out few column 
X_train.head(5)
```

```python
Y_train.head(5)
```

```python
# dropping Unique ID as its not require in training 
X_train.drop('Unique_ID',inplace=True,axis=1)
X_test.drop('Unique_ID',inplace=True,axis=1)
Y_train = Y_train['Dependent_Variable']
```

### DATA CLEANING

```python
# plotting null value on heatmap
sns.heatmap(X_train.isnull(), cbar=False)
```

```python
sns.heatmap(X_test.isnull(), cbar=False)
```

```python
# checking % of null value for each column
X_train.isnull().mean()*100
```

```python
# column N25-N32 ---> 85% data missing therfore these columns are of no use
# dropping N25 - N32 columns from data set
```

```python
X_train.drop(columns=['N25','N26','N27','N28','N29','N30','N31','N32'],inplace=True,axis=1)
```

```python
X_test.drop(columns=['N25','N26','N27','N28','N29','N30','N31','N32'],inplace=True,axis=1)
```

```python
# replacing TRUE --> 1 , FALSE --> 0
X_train.replace({False: 0, True: 1}, inplace=True)
X_test.replace({False: 0, True: 1}, inplace=True)
```

```python
X_train.head(5)
```

### filling remaining column with mean-values using simple imputer

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X_train)
```

```python
X_trf_train=imputer.transform(X_train)
```

```python
X_trf_train = pd.DataFrame(X_trf_train , columns=X_train.columns)
```

```python
X_trf_train.head(5)
```

```python
imputer_X_test = SimpleImputer(missing_values=np.nan,strategy='mean')
imputer_X_test.fit(X_test)
X_trf_test = imputer_X_test.transform(X_test)
X_trf_test = pd.DataFrame(X_trf_test , columns=X_test.columns)
```

```python
X_trf_test.head(5)
```

```python
# no null value present in train and test data..
sns.heatmap(X_trf_train.isnull(), cbar=False)
```

### CREATING ANDTRAINING THE MODEL

```python
X_trf_train.shape
```

```python
# creating Nueral network model
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(11,activation='relu',input_dim=35))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#training the model
model.fit(X_trf_train,Y_train,epochs=300)
```

```python
Y_pred = model.predict(X_trf_test)
```

```python
Y_pred.shape
```

```python
# probablity of target class
Y_pred
```

```python
Y_pred = [(x[0]) for x in Y_pred]
```

### Writing final prediction in csv file

```python
test_data =pd.read_csv('Test/X_test.csv')
```

```python
final_predictions = pd.DataFrame({'Unique_ID':test_data["Unique_ID"],'Class_1_Probablity':Y_pred})
```
